```python
dados = pd.read_json('dadoslimpos.json')

#exibindo todas as colunas
pd.set_option('display.max_columns', None)
```

É exibido a distribuição das notas em uma disciplina a partir do gráfico violino (https://en.wikipedia.org/wiki/Violin_plot) e realizado o teste estatístico (https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.probplot.html) se a distribuição se comporta como uma Normal.

A partir dos gráficos abaixo é nitidamente verificado que as notas não seguem uma distribuição normal, mas, aparentemente, uma distribuição bimodal. 

```python
#gerando graficos
%matplotlib inline

plt.close('all')

cnt = 0

#loop sobre cada disciplina
for index1, group1 in dados.groupby(['descricao_historico']):    
    fig = plt.figure()
    sns.plt.suptitle(index1)
    sns.violinplot(dados[dados.descricao_historico == index1]['media_final'])
    
    fig = plt.figure()
    stats.probplot(dados[dados.descricao_historico == index1]['media_final'], plot=plt)    
    
    cnt += 1
    if cnt == 5:
        break
```

É realizado a clusterização por disciplina em dois grupos (provavelmente alunos aprovado e reprovados). 
É criado dentro do frame um novo atributo chamado 'classe', o qual identifica a qual grupo aquele aluno pertence. 

A partir dos gráficos gerados é verificado que a classe que, aparentemente, representa os alunos aprovados está bem modelada pela distribuição normal. Entretanto, a classe que, aparentemente, representa os alunos reprovados não apresenta uma boa modelagem. Tal fenômeno talvez possa ser interpretado que há 2 grupos dentro dos alunos não aprovados: os que evadem (com uma grande quantidade de zeros) e os que tentam até o final a disciplina, mas não conseguem êxito. 

```python
%matplotlib inline
plt.close('all')

#cria coluna 'classe' no frame dados inicializado com -1, a fim de controle
dados['classe'] = np.ones(dados.shape[0])*-1

#definir linha que divide os clusters
cnt = 0

for index1, group1 in dados.groupby(['descricao_historico']):    
    
    #recuperando as notas apenas daquela disciplina    
    data = dados[dados.descricao_historico == index1]['media_final']
    
    #restrição mínima de 10 instâncias  
    if data.count() < 10:
        continue
    
    #recuperando os clusters
    cluster = KMeans(n_clusters=2, init=np.array([[0], [10]]))
    cluster.fit(data.reshape(-1,1))
    #print(cluster.cluster_centers_)
        
    #atribuindo a classe
    dados.loc[dados.descricao_historico == index1, 'classe'] = cluster.predict(data.reshape(-1,1))
    
    #figura mostrando a clusterizacao
    fig = plt.figure()
    sns.plt.suptitle(index1)
    sns.distplot( dados[(dados.descricao_historico == index1) & (dados.classe == 0)]['media_final'] )
    sns.distplot( dados[(dados.descricao_historico == index1) & (dados.classe == 1)]['media_final'] )
    plt.plot(cluster.cluster_centers_, [0, 0], "o")
    
    #figura mostrando o ajuste dos dados aos clusters
    fig = plt.figure()
    stats.probplot(dados[(dados.descricao_historico == index1) & (dados.classe == 0)]['media_final'], plot=plt)
    stats.probplot(dados[(dados.descricao_historico == index1) & (dados.classe == 1)]['media_final'], plot=plt)
    
    cnt += 1
    if cnt == 5:
        break
```

```python
%matplotlib inline
plt.close('all')

#cria coluna 'classe' no frame dados inicializado com -1, a fim de controle
dados['classe'] = np.ones(dados.shape[0])*-1

#definir linha que divide os clusters
cnt = 0

for index1, group1 in dados.groupby(['descricao']):    
    
    #recuperando as notas apenas daquela disciplina    
    data = dados[dados.descricao == index1]['media_final']
    
    #recuperando os clusters
    cluster = KMeans(n_clusters=3, init=np.array([[0], [5], [10]]))
    cluster.fit(data.reshape(-1,1))
    print(cluster.cluster_centers_)
        
    #atribuindo a classe
    dados.loc[dados.descricao == index1, 'classe'] = cluster.predict(data.reshape(-1,1))
    
    #figura mostrando a clusterizacao
    fig = plt.figure()
    sns.distplot( dados[(dados.descricao == index1) & (dados.classe == 0)]['media_final'] )
    sns.distplot( dados[(dados.descricao == index1) & (dados.classe == 1)]['media_final'] )
    sns.distplot( dados[(dados.descricao == index1) & (dados.classe == 2)]['media_final'] )
    plt.plot(cluster.cluster_centers_, [0, 0, 0], "o")
    
    #figura mostrando o ajuste dos dados aos clusters
    fig = plt.figure()
    stats.probplot(dados[(dados.descricao == index1) & (dados.classe == 0)]['media_final'], plot=plt)
    stats.probplot(dados[(dados.descricao == index1) & (dados.classe == 1)]['media_final'], plot=plt)
    stats.probplot(dados[(dados.descricao == index1) & (dados.classe == 2)]['media_final'], plot=plt)
    
    cnt += 1
    if cnt == 10:
        break
```
